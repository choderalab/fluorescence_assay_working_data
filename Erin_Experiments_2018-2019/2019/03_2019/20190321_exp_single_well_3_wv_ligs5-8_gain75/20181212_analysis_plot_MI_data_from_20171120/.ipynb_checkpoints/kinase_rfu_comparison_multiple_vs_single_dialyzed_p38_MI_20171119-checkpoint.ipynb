```python
# %load kinase_multiple_well_spectra_experiment


# get_ipython().run_line_magic('matplotlib', 'inline')
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.offsetbox import AnchoredText
import matplotlib.cm as cm
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import FormatStrFormatter
import matplotlib
from matplotlib import gridspec
from glob import glob


# In[13]:


import assaytools


# In[14]:


from assaytools import platereader


# In[15]:

filenames = ['2017-11-20 15-48-06_plate_1.xml',
            '2017-11-20 16-17-59_plate_1.xml',
            '2017-11-20 16-41-35_plate_1.xml',
            '2017-11-20 16-59-09_plate_1.xml',
            '2017-11-20 17-19-31_plate_1.xml',
            '2017-11-20 17-37-37_plate_1.xml',
            '2017-11-20 17-58-04_plate_1.xml',
            '2017-11-20 18-16-02_plate_1.xml',
            '2017-11-20 18-35-42_plate_1.xml',
            '2017-11-20 18-54-33_plate_1.xml',
            '2017-11-20 19-12-49_plate_1.xml',
            '2017-11-20 19-31-32_plate_1.xml']

datapath = 'infinite_results/single_wv'


# In[16]:


concentrations = np.array([  0.00000000e+00,   8.00000000e-09,   1.75e-08,
         3.83e-08,   8.37e-08,   1.83e-07,
         4e-07,   8.75e-07,   1.91e-06,
         4.18e-06,   9.15e-06,   2e-05])


# In[17]:

reads_multi = list()
nreads_multi = len(filenames_multi)
for read in range(nreads_multi):
    read = platereader.read_icontrol_xml(os.path.join(datapath_multi, filenames_multi[read]))
    reads_multi.append(read)
    
label_multi = 'ex280_scan_top_gain100'
```

## Get data from single well

```python
# %load kinase_single_well_spectra_experiment

filenames_single = ['p38_Abl_GK_WT_Src_conc_0_20190213_111154.xml',
             'p38_Abl_GK_WT_Src_conc_1_20190213_112906.xml',
             'p38_Abl_GK_WT_Src_conc_2_20190213_114438.xml',
             'p38_Abl_GK_WT_Src_conc_3_20190213_120010.xml',
             'p38_Abl_GK_WT_Src_conc_4_20190213_121543.xml',
             'p38_Abl_GK_WT_Src_conc_5_20190213_123117.xml',
             'p38_Abl_GK_WT_Src_conc_6_20190213_124651.xml',
             'p38_Abl_GK_WT_Src_conc_7_20190213_130224.xml',
             'p38_Abl_GK_WT_Src_conc_8_20190213_131758.xml',
             'p38_Abl_GK_WT_Src_conc_9_20190213_133335.xml',
             'p38_Abl_GK_WT_Src_conc_10_20190213_134909.xml',
             'p38_Abl_GK_WT_Src_conc_11_20190213_140442.xml',
             'p38_Abl_GK_WT_Src_conc_12_20190213_142015.xml',
             'p38_Abl_GK_WT_Src_conc_13_20190213_143547.xml',
             'p38_Abl_GK_WT_Src_conc_14_20190213_145126.xml',
             'p38_Abl_GK_WT_Src_conc_15_20190213_150700.xml',
             'p38_Abl_GK_WT_Src_conc_16_20190213_152237.xml']

datapath_single = 'infinite_results'


# In[16]:


concentrations_single = np.array([  0.00000000e+00,   8.00000000e-09,   1.34778097e-08,
         2.27064194e-08,   3.82541000e-08,   6.44476851e-08,
         1.08576705e-07,   1.82922021e-07,   3.08173524e-07,
         5.19188015e-07,   8.74689659e-07,   1.47361260e-06, 2.48263378e-06,
         4.18255821e-06, 7.04646547e-06, 1.118713651e-05, 2.0e-05])


# In[17]:

reads_single = list()
nreads_single = len(filenames_single)
for read in range(nreads_single):
    read = platereader.read_icontrol_xml(os.path.join(datapath_single, filenames_single[read]))
    reads_single.append(read)


# In[18]:


plate_layout = pd.read_excel('20190213_plate_layout_for_binding_assay.xlsx', header=None)
plate_layout = plate_layout.loc[:, 0:1]
plate_layout.columns = ['well', 'name']
if len(list(plate_layout[plate_layout.well == 'A1'].name)[0].split('_')) > 0:
    list(plate_layout[plate_layout.well == 'A1'].name)[0].split('_')
    
label_single = 'ex280_scan_top_gain100'
```

### Plot data at all wavelengths

```python
### extract data for single well, break up into each ligand to mirror multiple well data extraction 

# get protein data

measurements_single_prot = list()
prot_well_dict = {'Bos':'B8', 'Bos_iso':'D8', 'Erl':'F8', 'Gef':'H8'}
for key, value in prot_well_dict.items():
    well = value
    x_values = list(reads_single[0][label_single][well].keys())
    measurements = [list(reads_single[i][label_single][well].values())[1] for i in range(nreads_single)]
    for i in range(len(measurements)):
        measurements = list([x if x != 'OVER' else 0 for x in measurements])
        measurements = [float(i) for i in measurements]
    measurements_single_prot.append(measurements)

    
# get buffer data
measurements_single_buff = list()
buff_well_dict = {'Bos':'A2', 'Bos_iso':'C2', 'Erl':'E2', 'Gef':'G2'}
for key, value in buff_well_dict.items():
    well = value
    x_values = list(reads_single[0][label_single][well].keys())
    measurements = [list(reads_single[i][label_single][well].values())[1] for i in range(nreads_single)]
    for i in range(len(measurements)):
        measurements = list([x if x != 'OVER' else 0 for x in measurements])
        measurements = [float(i) for i in measurements]
    measurements_single_buff.append(measurements)
```

### Plot data at 480 nm

```python
## Name wells and assign corresponding measurements at 480 nm based on ligand placement in plate layout

Bos_prot = np.zeros(12)
Bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'A%s' % index

    Bos_prot[well] = reads_multi[0]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'B%s' % index
    Bos_buff[well] = reads_multi[0]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Bos_iso_prot = np.zeros(12)
Bos_iso_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'C%s' % index

    Bos_iso_prot[well] = reads_multi[1]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'D%s' % index
    Bos_iso_buff[well] = reads_multi[1]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Erl_prot = np.zeros(12)
Erl_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'E%s' % index
    Erl_prot[well] = reads_multi[2]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'F%s' % index
    Erl_buff[well] = reads_multi[2]['ex280_scan_top_gain100'][buff_well_name]['480']

Gef_prot = np.zeros(12)
Gef_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'G%s' % index
    Gef_prot[well] = reads_multi[3]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'H%s' % index
    Gef_buff[well] = reads_multi[3]['ex280_scan_top_gain100'][buff_well_name]['480']
```

# Logx plot

```python
## make 4 plots, one for each ligand at 480 nm Em
current_palette = sns.color_palette("hls", 8)

wavelength = '480'
protein_color = current_palette[5]
protein_color_single = current_palette[4]
buffer_color = 'gray'

fig = plt.figure(figsize=(10,8), dpi=300)
plt.suptitle('Fluorescence at 480 nm (Dialyzed WT Src)', fontsize=25, fontweight='bold')
ax1 = fig.add_subplot(2,2,1)

# Plot fluorescence
plt.loglog(concentrations_multi, np.flip(Bos_prot, axis=0), c=protein_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_multi, np.flip(Bos_buff, axis=0), c=buffer_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_prot[0], c=protein_color_single, marker='s', linestyle='None', markersize=6, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_buff[0], c=buffer_color, marker='s', linestyle='None', markersize=6, alpha=0.6)
ax1.set_ylim(10**2, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_ticks([])
plt.minorticks_off()
plt.legend(['Multiple Well: WT Src + ligand', 'Multiple Well: ligand in buffer', 'Single Well: WT Src + ligand', 'Single Well: ligand in buffer'], loc=2, prop={'size': 8})
plt.title('Bosutinib')

ax1 = fig.add_subplot(2,2,2)


# Plot fluorescence
plt.loglog(concentrations_multi, np.flip(Bos_iso_prot, axis=0), c=protein_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_multi, np.flip(Bos_iso_buff, axis=0), c=buffer_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_prot[1], c=protein_color_single, marker='s', linestyle='None', markersize=6, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_buff[1], c=buffer_color, marker='s', linestyle='None', markersize=6, alpha=0.6)
ax1.set_ylim(10**2, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
plt.minorticks_off()
plt.legend(['Multiple Well: WT Src + ligand', 'Multiple Well: ligand in buffer', 'Single Well: WT Src + ligand', 'Single Well: ligand in buffer'], loc=2, prop={'size': 8})
plt.title('Bosutinib Isomer')

ax1 = fig.add_subplot(2,2,3)

# Plot fluorescence
plt.loglog(concentrations_multi, np.flip(Erl_prot, axis=0), c=protein_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_multi, np.flip(Erl_buff, axis=0), c=buffer_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_prot[2], c=protein_color_single, marker='s', linestyle='None', markersize=6, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_buff[2], c=buffer_color, marker='s', linestyle='None', markersize=6, alpha=0.6)
ax1.set_ylim(10**2, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_tick_params(labelsize=15)
plt.legend(['Multiple Well: WT Src + ligand', 'Multiple Well: ligand in buffer', 'Single Well: WT Src + ligand', 'Single Well: ligand in buffer'], loc=2, prop={'size': 8})
plt.title('Erlotinib HCl')
axis = plt.gca()
plt.ylabel('Fluorescence Units', fontsize=25, fontweight='bold')
axis.yaxis.set_label_coords(-0.25,1.25)
plt.xlabel('Concentration (M)', fontsize=25, fontweight='bold')
axis.xaxis.set_label_coords(1,-0.25)
    

# Plot fluorescence
ax1 = fig.add_subplot(2,2,4)
plt.loglog(concentrations_multi, np.flip(Gef_prot, axis=0), c=protein_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_multi, np.flip(Gef_buff, axis=0), c=buffer_color, marker='.', linestyle='None', markersize=12, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_prot[3], c=protein_color_single, marker='s', linestyle='None', markersize=6, alpha=0.6)
plt.loglog(concentrations_single, measurements_single_buff[3], c=buffer_color, marker='s', linestyle='None', markersize=6, alpha=0.6)
ax1.set_ylim(10**2, 100000)
ax1.yaxis.set_ticks([])

ax1.xaxis.set_tick_params(labelsize=15)
plt.legend(['Multiple Well: WT Src + ligand', 'Multiple Well: ligand in buffer', 'Single Well: WT Src + ligand', 'Single Well: ligand in buffer'], loc=2, prop={'size': 8})
plt.title('Gefitinib')
    


fig.savefig('20190220_rfu_comparison_multi_vs_single_WT Src_dialyzed_logx_%s.pdf' % wavelength, dpi=300)
```
