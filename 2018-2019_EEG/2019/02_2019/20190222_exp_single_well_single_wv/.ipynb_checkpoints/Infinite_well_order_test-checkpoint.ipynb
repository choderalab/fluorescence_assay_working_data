```python
nrows = 8
ncols = 12
index = 0

well_order = list()
for row_index in range(nrows):
    for col_index in range(ncols):
        row = '%c' % (ord('A')+row_index)
        col = '%d' % (col_index+1)
        well = row + col
        well_order.append(well)

print(well_order)
```
