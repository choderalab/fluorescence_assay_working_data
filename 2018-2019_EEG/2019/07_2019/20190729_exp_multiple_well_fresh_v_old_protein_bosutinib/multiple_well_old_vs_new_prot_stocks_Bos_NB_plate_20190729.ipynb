```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.offsetbox import AnchoredText
import matplotlib.cm as cm
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import FormatStrFormatter
import matplotlib
from matplotlib import gridspec
from glob import glob

import assaytools
from assaytools import platereader
```

```python
filenames = sorted(glob('infinite_results/*.xml'))
filenames.sort(key=os.path.getmtime)

reads = list()
nreads = len(filenames)
for read in range(nreads):
    read = platereader.read_icontrol_xml(filenames[read])
    reads.append(read)
    
label = 'ex280_scan_top_gain100'
```

```python
concentrations = np.array([  0.00000000e+00,   8.00000000e-09,   1.75e-08,
         3.83e-08,   8.37e-08,   1.83e-07,
         4e-07,   8.75e-07,   1.91e-06,
         4.18e-06,   9.15e-06,   2e-05])
```

### Plot data at 480 nm

```python
## Name wells and assign corresponding measurements at 480 nm based on ligand placement in plate layout

Abl_old_Bos_prot = np.zeros(12)
Abl_old_Bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'A%s' % index

    Abl_old_Bos_prot[well] = reads[0]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'B%s' % index
    Abl_old_Bos_buff[well] = reads[0]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Abl_new_Bos_prot = np.zeros(12)
Abl_new_Bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'C%s' % index

    Abl_new_Bos_prot[well] = reads[1]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'D%s' % index
    Abl_new_Bos_buff[well] = reads[1]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Src_old_Bos_prot = np.zeros(12)
Src_old_Bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'E%s' % index
    Src_old_Bos_prot[well] = reads[2]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'F%s' % index
    Src_old_Bos_buff[well] = reads[2]['ex280_scan_top_gain100'][buff_well_name]['480']

Src_new_Bos_prot = np.zeros(12)
Src_new_Bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'G%s' % index
    Src_new_Bos_prot[well] = reads[3]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'H%s' % index
    Src_new_Bos_buff[well] = reads[3]['ex280_scan_top_gain100'][buff_well_name]['480']
```

# Logx plot

```python
## make 4 plots, one for each ligand at 480 nm Em
current_palette = sns.color_palette("husl", 10)

wavelength = '480'
protein_color = current_palette[6]
buffer_color = 'gray'

fig = plt.figure(figsize=(10,8), dpi=300)
plt.suptitle('Fluorescence at 480 nm', fontsize=18, fontweight='bold')
ax1 = fig.add_subplot(2,1,1)

plt.semilogx(concentrations, np.flip(Abl_old_Bos_prot, axis=0), c=protein_color, marker='.', linestyle='-', markersize=12)
plt.semilogx(concentrations, np.flip(Abl_old_Bos_buff, axis=0), c=buffer_color, linestyle='--', markersize=12)
plt.semilogx(concentrations, np.flip(Abl_new_Bos_prot, axis=0), c=protein_color, marker='s', linestyle='-', markersize=8)
plt.semilogx(concentrations, np.flip(Abl_new_Bos_buff, axis=0), c=buffer_color, linestyle='-.', markersize=12)
ax1.set_ylim(0, 60000)

ax1.yaxis.set_tick_params(labelsize=10)
ax1.xaxis.set_tick_params(labelsize=10)
plt.minorticks_off()
plt.legend(['Abl WT old + ligand', 'ligand', 'Abl WT new + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Abl WT')

ax1 = fig.add_subplot(2,1,2)

# Plot fluorescence
plt.semilogx(concentrations, np.flip(Src_old_Bos_prot, axis=0), c=protein_color, marker='.', linestyle='-', markersize=12)
plt.semilogx(concentrations, np.flip(Src_old_Bos_buff, axis=0), c=buffer_color, linestyle='--', markersize=12)
plt.semilogx(concentrations, np.flip(Src_new_Bos_prot, axis=0), c=protein_color, marker='s', linestyle='-', markersize=8)
plt.semilogx(concentrations, np.flip(Src_new_Bos_buff, axis=0), c=buffer_color, linestyle='-.', markersize=12)
ax1.set_ylim(0, 60000)

ax1.yaxis.set_tick_params(labelsize=10)
ax1.xaxis.set_tick_params(labelsize=10)
plt.minorticks_off()
plt.legend(['Src WT old + ligand', 'ligand', 'Src WT new + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Src WT')

plt.ylabel('Fluorescence Units', fontsize=18, fontweight='bold')
ax1.yaxis.set_label_coords(-0.1,1.25)
plt.xlabel('[Bosutinib] (M)', fontsize=18, fontweight='bold')
ax1.xaxis.set_label_coords(0.5,-0.15)

fig.savefig('20190729_old_v_new_prot_NBS_plate_logx_%s.pdf' % wavelength, dpi=300)
```
