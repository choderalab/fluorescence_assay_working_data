```python
# %load kinase_single_well_spectra_experiment


# get_ipython().run_line_magic('matplotlib', 'inline')
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.offsetbox import AnchoredText
import matplotlib.cm as cm
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import FormatStrFormatter
import matplotlib
from glob import glob


# In[13]:


import assaytools


# In[14]:


from assaytools import platereader


# In[15]:

# filenames = glob('*.xml')
# filenames = str(filenames.sort())

filenames = ['EEG_fluo_test_20181115_122759.xml']

datapath = 'infinite_results'


# In[16]:


concentrations = np.array([0.00000000e+00, 2.0e-05])


# In[17]:

reads = list()
nreads = len(filenames)
for read in range(nreads):
    read = platereader.read_icontrol_xml(os.path.join(datapath, filenames[read]))
    reads.append(read)


# In[18]:


plate_layout = pd.read_excel('20181114_plate_layout_for_troubleshooting_assay.xlsx', header=None)
plate_layout = plate_layout.loc[:, 0:1]
plate_layout.columns = ['well', 'name']
if len(list(plate_layout[plate_layout.well == 'A1'].name)[0].split('_')) > 0:
    list(plate_layout[plate_layout.well == 'A1'].name)[0].split('_')
```

```python
# extract reads from xml file

nrows = 5
ncols = 12
index = 0

label = 'ex280_scan_top_gain100_Copy2'

title = 'Specta Reads at 280nm'
data = np.zeros([nrows, ncols, 65], np.float64)
for row_index in range(nrows):
    for col_index in range(ncols):
        row = '%c' % (ord('A')+row_index)
        col = '%d' % (col_index+1)
        well = row + col
        x_values = list(reads[0][label][well].keys())
        measurements = list(reads[0][label][well].values())
        measurements = [x if x != 'OVER' else 0 for x in measurements]
        data[row_index,col_index,:] = measurements
    
# extract buffer reads from xml file (ncols only goes up to F10 so ncols = 12 won't work here)

nrows2 = 1
ncols2 = 10
index2 = 0
data2 = np.zeros([nrows2, ncols2, 65], np.float64)
for row_index in range(nrows2):
    for col_index in range(ncols2):
        row2 = '%c' % (ord('F')+row_index)
        col2 = '%d' % (col_index+1)
        well2 = row2 + col2
        measurements2 = list(reads[0][label][well2].values())
        measurements2 = [x if x != 'OVER' else 0 for x in measurements2]
        data2[row_index,col_index,:] = measurements2
        
# c = data2[row_index,col_index,:]
# print(list(c))
```

# Plot fluorescence across all 65 wavelengths

```python
# plot fluorescence across 65 wavelengths

nrows = 5
ncols = 12
index = 0
fig, axes = plt.subplots(6, ncols, figsize=[30,25])


label = 'ex280_scan_top_gain100_Copy2'
title = 'Fluorescence Spectra at Ex 280nm (fluo test)'


ymax = data.max().max().max()
xmax = max(list(reads[0][label][well].keys()))
xmin = min(list(reads[0][label][well].keys()))
c = list(concentrations)

# make array for color coding the plots with 0 uM ligand gray
buff_wells = ['A1', 'A3', 'A5', 'A7', 'A9', 'A11', 'B1', 'B3', 'B5', 'B7', 'B9', 'B11', 'C1', 'C3', 'C5', 'C7', 'C9', 'C11', 'D1', 'D3', 'D5', 'D7', 'D9', 'D11', 'E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10','E11', 'E12']

for row_index in range(nrows):
    for col_index in range(ncols):
        row = '%c' % (ord('A')+row_index)
        col = '%d' % (col_index+1)
        well = row + col
        if well in buff_wells:
            color = '0.75'
        else:
            color = 'b'
        if len(list(plate_layout[plate_layout.well == well].name)) > 0:
            ax = plt.subplot(6,ncols,index+1)
            plt.semilogy(x_values, list(data[row_index, col_index,:]), c= color, linewidth=2);
                

            #if index == 36:
            #    plt.xticks(np.arange(int(x_values[20]), 600, 40),x_values[20::8], rotation=90, fontsize=18)

            plt.xticks([]);
            plt.yticks([]);
            plt.ylim(0, 30000)
            # label subplots with contents of wells and index # 
            if len(list(plate_layout[plate_layout.well == well].name)) > 0:
                anchored_text = AnchoredText('%s' % list(plate_layout[plate_layout.well == well].name)[0].split('_')[0], loc=1, prop=dict(size=18))
                ax.add_artist(anchored_text)
            if index == 47:
                plt.suptitle(title, fontsize=70, fontweight='bold')
            if index == 49:
                axis = plt.gca()
                plt.ylabel('Log Fluorescence Units (max 30,000)', fontsize=55, fontweight='bold')
                axis.yaxis.set_label_coords(-2,2.5)
            if index == 1 or index == 3 or index == 5 or index == 7 or index == 9 or index == 11:
                plt.title('20 uM', loc='center', fontsize=20)
            if index == 0 or index == 2 or index == 4 or index == 6 or index == 8 or index == 10:
                plt.title('0 uM', loc='center', fontsize=20)
                
            # ligand labels
            if index == 0:
                plt.ylabel('Bosutinib', fontsize=20)
            if index == 12:
                plt.ylabel('Bosutinib Isomer', fontsize=20)
            if index == 24:
                plt.ylabel('Erlotinib HCl', fontsize=20)
            if index == 36:
                plt.ylabel('Gefitinib', fontsize=20)
            if index == 48:
                plt.ylabel('Protein & Buffer', fontsize=16)
                
            index += 1
            anchored_text2 = AnchoredText('%s' % (index),  loc=3, prop=dict(size=14))
            ax.add_artist(anchored_text2)

# add plots for ligand in buffer
nrows2 = 1
ncols2 = 10

for row_index in range(nrows2):
    for col_index in range(ncols2):
        row2 = '%c' % (ord('F')+row_index)
        col2 = '%d' % (col_index+1)
        well2 = row2 + col2
        if well2 in buff_wells:
            color = '0.75'
        else:
            color = 'b'
        if len(list(plate_layout[plate_layout.well == well].name)) > 0:
            ax2 = plt.subplot(nrows2+5,ncols,index+1)
            plt.semilogy(x_values, list(data2[row_index, col_index,:]), c=color, linewidth=2);
            
            if len(list(plate_layout[plate_layout.well == well2].name)) > 0:
                anchored_text2 = AnchoredText('%s' % list(plate_layout[plate_layout.well == well2].name)[0].split('_')[0], loc=1, prop=dict(size=16))
                ax2.add_artist(anchored_text2)
            
            plt.xticks([]);
            plt.yticks([]);
            plt.ylim(0, 30000)
            
            if index == 60:
                plt.ylabel('Ligand & Buffer', fontsize=16)
            if index == 65:
                plt.xlabel('Wavelength (nm)', fontsize=75, fontweight='bold')  
                
            index += 1
            anchored_text2 = AnchoredText('%s' % (index),  loc=3, prop=dict(size=14))
            ax2.add_artist(anchored_text2)

            # hide last two subplots, which aren't being used (F11 & F12)
            axes[-1, -1].axis('off')
            axes[-1,-2].axis('off')
            
fig.savefig('20181128_test_single_well_spectra_log_y.pdf', bbox_inches='tight', dpi=300)


```
