## Normalize data for better comparison

### normalization equation: normalized_flu = [(flu_in_well_i) - (min_flu)] / [(max_flu - min_flu)]

```python
# get min and max flu for each well then normalize data points (02/28/2019 data)

nrows = 8
ncols = 12

min_flu_20190228 = np.zeros([nrows, ncols], np.float64)
max_flu_20190228 = np.zeros([nrows, ncols], np.float64)
data_norm_20190228 = np.zeros([nrows, ncols, nreads], np.float64)

for row_index in range(nrows):
    for col_index in range(ncols):
        row = '%c' % (ord('A')+row_index)
        col = '%d' % (col_index+1)
        well = row + col
        measurements_20190228 = [(reads[i][label][well]) for i in range(nreads)]
        for i in range(nreads):
            measurements_20190228 = [x if x != 'OVER' else 0 for x in measurements_20190228]
            measurements_20190228 = [float(i) for i in measurements_20190228]
            
        max_flu_20190228[row_index,col_index] = max(measurements_20190228)
        min_flu_20190228[row_index,col_index] = min(measurements_20190228)
        
        data_norm_20190228[row_index,col_index,:] = np.array([((i-min_flu_20190228[row_index,col_index])/(max_flu_20190228[row_index,col_index]-min_flu_20190228[row_index,col_index])) for i in measurements_20190228])
        
        print(well)
        print(data_norm_20190228[row_index,col_index])
```
